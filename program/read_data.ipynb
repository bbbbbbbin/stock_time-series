```python
import csv
import xlrd
import json
import os
from tqdm import tqdm
```

```python
data_path = 'data/'
```

```python
data_list = os.listdir(data_path)
data_list.sort(key = lambda x: int(x[:-4]))
```

```python
def readData(num,filename,dic):
    # print(filename)
    csvfile = open(filename, encoding = 'utf-8')
    datas = csv.reader(csvfile)
    for data in datas:
        if data[0] == 'code':
            continue
        if data[0] not in dic:
            dic[data[0]]={}
        dic[data[0]][num] = data[1]
    return dic
```

```python
dic = {}
num = 1
for i in tqdm(range(0,len(data_list))):
    path = os.path.join(data_path,data_list[i])
    # print(path)
    if os.path.isfile(path):
        dic = readData(num,path,dic)
        num += 1
```

```python
with open('rounda.json','w') as f:
    f.write(json.dumps(dic,ensure_ascii=False,indent=2))
    f.close()
```
