```python
x = np.array(list(dicstock102419.keys()),dtype='float64')
y = np.array(list(dicstock102419.values()),dtype='float64')
y = np.log(y)
data = pd.DataFrame(y,x)
plt.figure(figsize=(16,9))
sns.lineplot(x,y)
da, db = sm.tsa.filters.hpfilter(y)
```

```python
data1 = data[0].diff(1).dropna()
plt.figure(figsize=(16,9))
sns.lineplot(x[1:],data1)
dftest = adfuller(data1)
print(dftest)
```

```python
m = 10
data = da
acf,q,p = sm.tsa.acf(data,nlags=m,qstat=True)
pacf = sm.tsa.acf(data,nlags=m)
out = np.c_[range(1,m+1), acf[1:], pacf[1:], q, p]
output=pd.DataFrame(out, columns=['lag', "AC", "PAC", "Q", "P-value"])
output
```

```python
fig = plt.figure(figsize=(16,9))
ax1 = fig.add_subplot(211)
fig = sm.graphics.tsa.plot_acf(data,ax=ax1)
ax2 = fig.add_subplot(212)
fig = sm.graphics.tsa.plot_pacf(data,ax=ax2)
```

```python
from statsmodels.tsa.arima_model import ARMA
from statsmodels.tsa.stattools import q_stat
p = 2
q = 1
results_ARMA = ARMA(data, order=(p, q)).fit(disp=False)
print(results_ARMA.params)
sm.stats.durbin_watson(results_ARMA.resid)
```
